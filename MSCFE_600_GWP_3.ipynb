# **Step 5: Implementation**
**Python Code for Lending Money at a Fixed Rate for an Unsecured Purchase (e.g., Credit Card)**

```python
import random

class LoanApplication:
    def __init__(self, applicant_name, credit_score, loan_amount, co_signer_credit_score=None):
        self.applicant_name = applicant_name
        self.credit_score = credit_score
        self.loan_amount = loan_amount
        self.co_signer_credit_score = co_signer_credit_score
        self.approved = False
        self.loan_terms = None

    def assess_risk(self):
        # High risk if credit score is below 650
        if self.credit_score >= 650:
            return 'low'
        elif self.co_signer_credit_score and self.co_signer_credit_score >= 700:
            return 'medium'
        else:
            return 'high'

    def approve_loan(self):
        risk = self.assess_risk()
        if risk == 'low':
            self.approved = True
            self.loan_terms = self.generate_loan_terms(risk)
        elif risk == 'medium':
            self.approved = True
            self.loan_terms = self.generate_loan_terms(risk)
        else:
            self.approved = False

    def generate_loan_terms(self, risk):
        interest_rate = {
            'low': 0.05,
            'medium': 0.1,
            'high': 0.2
        }[risk]
        return {
            'interest_rate': interest_rate,
            'repayment_period_months': 36,
            'automatic_withdrawal': True
        }

def simulate_loan_approval(applicant_name, credit_score, loan_amount, co_signer_credit_score=None):
    application = LoanApplication(applicant_name, credit_score, loan_amount, co_signer_credit_score)
    application.approve_loan()
    return application

# Example usage
applicant = simulate_loan_approval("Rashedul Islam", 630, 10000, co_signer_credit_score=720)
if applicant.approved:
    print(f"Loan Approved for {applicant.applicant_name}")
    print("Loan Terms:", applicant.loan_terms)
else:
    print(f"Loan Denied for {applicant.applicant_name}")
```

# **Step 6: Implementation**
**Python Code for Lending Money with Security (e.g., Publicly Traded Equity)**

```python
import yfinance as yf

class SecuredLoanApplication:
    def __init__(self, applicant_name, security_ticker, loan_amount):
        self.applicant_name = applicant_name
        self.security_ticker = security_ticker
        self.loan_amount = loan_amount
        self.security_value = None
        self.loan_terms = None
        self.approved = False

    def assess_security_value(self):
        stock = yf.Ticker(self.security_ticker)
        self.security_value = stock.history(period='1d')['Close'][0]

    def approve_loan(self):
        self.assess_security_value()
        loan_to_value_ratio = self.loan_amount / self.security_value
        if loan_to_value_ratio <= 0.7:
            self.approved = True
            self.loan_terms = self.generate_loan_terms()
        else:
            self.approved = False

    def generate_loan_terms(self):
        return {
            'interest_rate': 0.05,
            'repayment_period_months': 24,
            'security': self.security_ticker,
            'loan_to_value_ratio': self.loan_amount / self.security_value
        }

def simulate_secured_loan_approval(applicant_name, security_ticker, loan_amount):
    application = SecuredLoanApplication(applicant_name, security_ticker, loan_amount)
    application.approve_loan()
    return application

# Example usage
applicant = simulate_secured_loan_approval("Rashedul Islam", "AAPL", 50000)
if applicant.approved:
    print(f"Loan Approved for {applicant.applicant_name}")
    print("Loan Terms:", applicant.loan_terms)
else:
    print(f"Loan Denied for {applicant.applicant_name}")
```
